```python
## Create a Week bucket an compare Weeks with another
```

```python
df = pd.read_csv(path + "Locations_00.csv")
df.head()
```

```python
day_old = 0
week_number = 0
for i in range(len(df)):
    
    day_new = df["Day"][i]%5
    # letzte größer gleich als aktuelles -> week number++
    if(day_new<day_old):
        week_number = week_number + 1
    day_old = df["Day"][i]%5
    df.loc[df.index == i, 'Week_num'] = week_number
    print("Index: "+ str(i)+ " Day: " + str(df["Day"][i])+ "  Week:" + str(week_number))
```

```python
df
```

```python
df.to_csv("loc_00_with_weeks.csv")
```

```python
corr = df[['Week_num', 'X', 'Y', 'Z']].corr()
# heatmap
ax = sns.heatmap(corr)
ax.set_title('Weeks')

plt.show()
```
